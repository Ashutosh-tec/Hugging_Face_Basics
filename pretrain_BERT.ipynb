## Preprocess

```python
import numpy as np
from datasets import Dataset
# Define a function to tokenize and chunk the data
def chunk_samples(batch):
    # Tokenize the batch, getting input IDs and attention masks
    tokenized_batch = tokenizer(batch) #, padding=False, truncation=False, return_tensors="np")

    # Concatenate all input IDs
    # input_ids_concat = np.concatenate(tokenized_batch['input_ids'])
    input_ids_concat = sum(tokenized_batch['input_ids'], [])

    # Chunk the concatenated input IDs into size of 512
    num_chunks = len(input_ids_concat) // 512
    chunks = [input_ids_concat[i*512:(i+1)*512] for i in range(num_chunks)]

    # Return the packed chunks
    return chunks

# Initialize an empty list to hold all the chunks
all_chunks = []

# Process the dataset in batches of 1000
batch_size = 1000
for i in range(0, len(dataset), batch_size):
    # Extract the batch text
    batch_text = dataset['text'][i:i+batch_size]

    # Tokenize and chunk the batch
    chunks = chunk_samples(batch_text)

    # Add the chunks to the list
    all_chunks.extend(chunks)

# Create a new dataset from the chunks
ds_chunked = Dataset.from_dict({'input_ids': all_chunks})

# Get the total number of new samples (chunks) in the resulting dataset
total_samples = len(ds_chunked['input_ids'])

print(f"Total number of samples (chunks) in the new dataset: {total_samples}")
```

```python
from transformers import DataCollatorForLanguageModeling
from torch.utils.data import DataLoader
from sklearn.model_selection import train_test_split
from datasets import Dataset, DatasetDict

# Split the dataset into training and test sets (5% test size)
train_dataset, test_dataset = ds_chunked.train_test_split(test_size=0.05, seed=42).values()

# Define the MLM data collator with a masking probability of 0.2
data_collator = DataCollatorForLanguageModeling(
    tokenizer=tokenizer,
    mlm=True,
    mlm_probability=0.2
)

# Convert the Hugging Face datasets into PyTorch tensors
train_dataset.set_format(type='torch', columns=['input_ids'])
test_dataset.set_format(type='torch', columns=['input_ids'])

# Use PyTorch DataLoader to load batches of data
train_dataloader = DataLoader(train_dataset, batch_size=8, shuffle=True, collate_fn=data_collator)
test_dataloader = DataLoader(test_dataset, batch_size=8, shuffle=False, collate_fn=data_collator)

# Fetch a batch of data from the train dataloader
batch = next(iter(train_dataloader))

# The input_ids after masking (collator applies masking)
masked_input_ids = batch['input_ids']

# Token IDs corresponding to unmasked tokens can be found as follows:
# If you want to inspect unmasked tokens (which remain unchanged), compare with original input
# In BERT, usually [MASK] token ID is 103, so you can find unmasked tokens where they are not equal to 103.
mask_token_id = tokenizer.mask_token_id
unmasked_tokens = masked_input_ids[masked_input_ids != mask_token_id]

# print("Masked Input IDs:\n", masked_input_ids)
# print("Unmasked Tokens:\n", unmasked_tokens)

# # If you want to get the token IDs corresponding to the unmasked tokens
# print(f"Unmasked token IDs: {unmasked_tokens.tolist()}")
```

```python
config = BertConfig(
    num_hidden_layers=6,
    hidden_size=384,
    intermediate_size=1536
)
# Step: Initialize the BERT model for MLM (Masked Language Modeling)
model = BertForMaskedLM(config)
```

```python
train_dataset = train_dataset.select(range(10000))
test_dataset = test_dataset.select(range(1000))
```

```python
from transformers import Trainer, TrainingArguments

# Define the training arguments
training_args = TrainingArguments(
    output_dir='./results',          # Output directory
    overwrite_output_dir=True,       # Overwrite the content of the output directory
    num_train_epochs=3,              # Number of training epochs
    per_device_train_batch_size=8,   # Batch size per device during training
    per_device_eval_batch_size=8,   # Batch size per device during evaluation
)

trainer = Trainer(
    model=model,
    args=training_args,
    data_collator=data_collator,
    train_dataset=train_dataset,
    eval_dataset=test_dataset,

)
```

```python
trainer.train()
```
